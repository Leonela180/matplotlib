1. Agrupa los siguientes datos en intervalos de amplitud 2:
2, 3, 5, 1, 6, 7, 8, 0, 8, 9, 6, 4, 4, 3, 2, 6, 3, 3, 2, 1, 1, 5, 6, 7, 4, 3, 2, 5, 4, 6
Construye una tabla de frecuencias completa y dibuja el histograma
correspondiente.

```python
datos = [2, 3, 5, 1, 6, 7, 8, 0, 8, 9, 6, 4, 4, 3, 2, 6, 3, 3, 2, 1, 1, 5, 6, 7, 4, 3, 2, 5, 4, 6]
bins = np.arange(0, 11, 2)
frecuencias, bordes = np.histogram(datos, bins=bins)
tabla_data = []
for i in range(len(frecuencias)):
    fila = {
        'Intervalo': f"[{bordes[i]}, {bordes[i+1]})",
        'Frecuencia': frecuencias[i],
        'Frecuencia Acumulada': sum(frecuencias[:i+1]),
        'Frecuencia Relativa': round(frecuencias[i]/len(datos), 2)
    }
    tabla_data.append(fila)
df = pd.DataFrame(tabla_data)
display(df)


plt.hist(datos, bins=bins, edgecolor='black', color='lightblue')
plt.title("Histograma de datos agrupados con línea de frecuencia")
plt.xlabel("Intervalos")
plt.ylabel("Frecuencia")
plt.grid(axis='y', linestyle='--', alpha=0.2)
plt.xticks(bins)
plt.show()

```

4. El número de móviles vendidos diariamente por un gran centro comercial,
corresponde a los siguientes datos:
18, 18, 18, 18, 18, 18, 18, 18, 19, 19, 19, 19, 19, 19, 19, 19, 19, 19, 20, 20, 20,
20, 20, 20, 20, 20, 20, 20, 20, 20, 21, 21, 21, 21, 21, 21, 22, 22, 22, 22
Crea una tabla de frecuencias y dibuja el diagrama de barras correspondiente.

```python
ventas = np.array([18, 18, 18, 18, 18, 18, 18, 18, 19, 19, 19, 19, 19, 19, 19, 19, 19, 19,
                   20, 20, 20, 20, 20, 20, 20, 20, 20, 20, 20,
                   21, 21, 21, 21, 21, 21,
                   22, 22, 22, 22])

# Valores únicos y frecuencias con numpy
valores, frecuencias = np.unique(ventas, return_counts=True)

# Construir tabla con lista de diccionarios
tabla_data = []
for v, f in zip(valores, frecuencias):
    tabla_data.append({
        'Número de móviles vendidos': v,
        'Frecuencia': f
    })

df = pd.DataFrame(tabla_data)

# Mostrar tabla
display(df)

# Graficar diagrama de barras
plt.bar(df['Número de móviles vendidos'], df['Frecuencia'], color='violet')
plt.title('Frecuencia de móviles vendidos diariamente')
plt.xlabel('Número de móviles vendidos')
plt.ylabel('Frecuencia')
plt.xticks(valores)
plt.grid(axis='y', linestyle='--', alpha=0.2)
plt.show()
```

5. El número de hermanos de los alumnos de una clase está reflejado en los
siguientes datos:
1, 2, 2, 3, 0, 1, 2, 1, 3, 0, 0, 2, 1, 1, 4, 3, 4, 2, 1, 4, 3, 2, 1, 0, 1, 0, 1, 1, 2, 1, 0, 3
Construye una tabla con frecuencias absolutas y porcentajes. Crea un diagrama
de sectores basándote en los porcentajes.

```python
hermanos = np.array([1, 2, 2, 3, 0, 1, 2, 1, 3, 0, 0, 2, 1, 1, 4, 3, 4, 2, 1, 4,
                     3, 2, 1, 0, 1, 0, 1, 1, 2, 1, 0, 3])

# Calcular frecuencias absolutas y porcentajes
valores, frecuencias = np.unique(hermanos, return_counts=True)
porcentajes = np.round(frecuencias / len(hermanos) * 100, 2)

# Crear tabla
tabla_data = []
for v, f, p in zip(valores, frecuencias, porcentajes):
    tabla_data.append({
        'Número de hermanos': v,
        'Frecuencia absoluta': f,
        'Porcentaje (%)': p
    })

df = pd.DataFrame(tabla_data)

# Mostrar tabla
display(df)

# Graficar diagrama de sectores
plt.figure(figsize=(7,7))
plt.pie(porcentajes, labels=valores, autopct='%1.1f%%')
plt.title('Distribución número de hermanos', color='grey', fontweight='bold', fontsize="20")
plt.show()
```

8. La tabla recoge la edad de un grupo de jóvenes encuestados:
o Edad (años): 15, 16, 17, 18, 19
o Frecuencia absoluta: 5, 8, 2, 20, 5
a) Realiza un diagrama de barras. b) Dibuja el polígono de frecuencias.


```python
df = pd.DataFrame({
    'Edad': ['15', '16', '17', '18', '19'],
    'Frecuencia absoluta': [5, 8, 2, 20, 5]
})

edades_numeric = df['Edad'].astype(int)
frecuencias = df['Frecuencia absoluta']

colors = ['#FFC0CB', '#FFC0CB', '#FFC0CB', '#8B008B', '#8B008B']  # tonos rosados y fucsia

fig, axes = plt.subplots(3, 1, figsize=(5, 7), constrained_layout=True)

# Gráfico 1: Diagrama de barras
axes[0].bar(edades_numeric, frecuencias, color=colors, edgecolor='black')
axes[0].set_title('Frecuencia de Edades (Diagrama de Barras)', fontsize=16, fontweight='bold', color='darkred')
axes[0].set_xlabel('Edad', fontsize=14)
axes[0].set_ylabel('Frecuencia absoluta', fontsize=14)
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# Gráfico 2: Polígono de frecuencias
axes[1].plot(edades_numeric, frecuencias, marker='o', linestyle='-', color='blue', linewidth=2)
axes[1].set_title('Polígono de Frecuencias', fontsize=16, fontweight='bold', color='navy')
axes[1].set_xlabel('Edad', fontsize=14)
axes[1].set_ylabel('Frecuencia absoluta', fontsize=14)
axes[1].set_xticks(edades_numeric)
axes[1].grid(True, linestyle='--', alpha=0.7)

# Gráfico 3: Diagrama de barras con línea de frecuencia encima
axes[2].bar(edades_numeric, frecuencias, color=colors, edgecolor='black', label='Frecuencia absoluta')
axes[2].plot(edades_numeric, frecuencias, color='navy', marker='o', linestyle='-', linewidth=2, label='Línea de frecuencia (polígono)')
axes[2].set_title('Barras + Polígono de Frecuencias', fontsize=16, fontweight='bold', color='darkgreen')
axes[2].set_xlabel('Edad', fontsize=14)
axes[2].set_ylabel('Frecuencia absoluta', fontsize=14)
axes[2].grid(axis='y', linestyle='--', alpha=0.7)
axes[2].legend()

plt.show()
```
